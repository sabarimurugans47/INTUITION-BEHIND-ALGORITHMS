# **What is Gradient Descent Optimization?**

*   The main goal of this optimization is to minimize the cost (error) and finding best fit line
*   There three types of gradient descent :

1.   Batch Gradient Descent
2.   Mini Batch Gradient Descent
3.   Stochastic Gradient Descent

*   This notebook contains practical implementation of these optimizers only for regression








# **a).Batch Gradient Descent**

*   It takes full dataset for computation




```python
import pandas as pd,numpy as np
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
X=np.arange(1,6)
Y = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)
```

```python
Data=pd.DataFrame({
    'X':X,
    'Y':Y.flatten()
})
```

```python
Data
```

# **Initializing random weights , bias , Learning rate**

```python
w=0.50
b=1.0
learning_rate=0.01
```

# **Iteration-1**

```python
from sklearn.metrics import mean_squared_error

def epoch_1():
    y_preds = []
    for x in X:
        y_preds.append(w * x + b)
    mse = mean_squared_error(Y.ravel(), y_preds)

    grad_w = (2 / len(X)) * sum(x * (w * x + b - y) for x, y in zip(X, Y.ravel()))
    grad_b = (2 / len(X)) * sum((w * x + b - y) for x, y in zip(X, Y.ravel()))

    new_w = w - learning_rate * grad_w
    new_b = b - learning_rate * grad_b

    print("MSE:", mse)
    print("Learning Rate:", learning_rate)
    print("New w:", new_w, ", New b:", new_b)
```

```python
epoch_1()
```

# **Iteration-2**

```python
def epoch_2():
  w=0.77
  b=1.07
  y_preds = []
  for x in X:
      y_preds.append(w * x + b)
  mse = mean_squared_error(Y.ravel(), y_preds)

  grad_w = (2 / len(X)) * sum(x * (w * x + b - y) for x, y in zip(X, Y.ravel()))
  grad_b = (2 / len(X)) * sum((w * x + b - y) for x, y in zip(X, Y.ravel()))

  new_w = w - learning_rate * grad_w
  new_b = b - learning_rate * grad_b

  print("MSE:", mse)
  print("Learning Rate:", learning_rate)
  print("New w:", new_w, ", New b:", new_b)
```

```python
epoch_2()
```

# **Loss is reduced from 16.75 to 9.89. Continue epoches until you get nearly same value between old mse and new mse**

# **What is convergence?**

*   Initially this optimization takes w,b randomly.
*   First it checks cost and how much points it correctly fitted on line.
*   Next it tries to reduce cost and make the line to fit most points
*   So finally at one stage , cost will reduce and most points fit on line
*   Further iterations didnt have difference after the final stage reaches
*   This is called convergence stage
