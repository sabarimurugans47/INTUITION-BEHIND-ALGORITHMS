## Step 2 : Compute entropy for features

```python
#sunny has 0 yes and 2 no
sunny_yes=0
sunny_no=2/2
log_sunny_yes=0
log_sunny_no=np.log2(sunny_no)
```

```python
sunny_yes*0,sunny_no*log_sunny_no
```

```python
entropy_sunny=-((0)+(0))
entropy_sunny
```

```python
#overcast has 2 yes and 0 no
overcast_yes=2/2
overcast_no=0
log_overcast_yes=np.log2(overcast_yes)
log_overcast_no=0
```

```python
overcast_yes*log_overcast_yes,overcast_no*log_overcast_no
```

```python
entropy_overcast=-((0)+(0))
entropy_overcast
```

```python
#rain has 2 yes and 1 no
rain_yes=2/3
rain_no=1/3
log_rain_yes=np.log2(rain_yes)
log_rain_no=np.log2(rain_no)
```

```python
rain_yes*log_rain_yes,rain_no*log_rain_no
```

```python
entropy_rain=-((-0.38997500048077083)+(-0.5283208335737187))
entropy_rain
```

## Step 3 : Compute weighted entropy

```python
weighted_entropy=(2/7)*0+(2/7)*0+(3/7)*entropy_rain
weighted_entropy
```

## Step 4 : Compute information gain

```python
info_gain=entropy_y-weighted_entropy
info_gain
```

## step 5 : Compute split info

```python
split_info=-(2/7)*np.log2(2/7)-(2/7)*np.log2(2/7)-(3/7)*np.log2(3/7)
split_info
```

## Step 6 : Compute Gain ratio

```python
gain_ratio=info_gain/split_info
gain_ratio
```

## Weather feature has gain ratio of 0.3800920111324277.Highest gain ratio will be considered as root node.
