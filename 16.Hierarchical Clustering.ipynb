### Compute pair-wise merging cost using Ward Linkage
=(|Ci|.|Cj| / |Ci| + |Cj| ).dist*2 |i,j|
* Ci - Initial Cluster
* Cj - Other cluster
* dist*2 - Square of euclidean distance
* i,j - Centroids

```python
#For sample 1 to all
A_B=(1.5-1.0)**2+(1.8-2.0)**2
A_C=(5.0-1.0)**2+(8.0-2.0)**2
A_D=(8.0-1.0)**2+(8.0-2.0)**2
A_E=(1.0-1.0)**2+(0.6-2.0)**2
print(A_B,A_C,A_D,A_E)
```

### Finalizing through Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
Z = linkage(df, method='ward')
plt.figure(figsize=(6, 4))
dendrogram(Z, labels=["A", "B", "C", "D", "E"])
plt.show()
```

### Other linkages for Agglomerative clustering

*   Single Linkage (uses minimum distance)

*   Average Linkage (Takes average)

### Divisive Analysis (DIANA)

```python
#Let's create a sample data
df=pd.DataFrame({
    'x':[1.0,1.5,5.0,8.0],
    'y':[2.0,1.8,8.0,8.0]
},index=['A','B','C','D'])
df
```

#### Compute Distance matrix using l2 (Euclidean)
image.png

```python
dist = np.sqrt(((df.values[:,None,:] - df.values[None,:,:])**2).sum(axis=2))
dist_df = pd.DataFrame(dist, index=df.index, columns=df.index)
print(dist_df)
```

#### Now compute pair-wise dissimilarities

```python
#for sample A (1/n*(a,b)+(a,c)+(a,d))
A=(1/3)*(0.538516+7.211103+9.219544)
#for sample B (1/n*(b,a)+(b,c)+(b,d))
B=(1/3)*(0.538516+7.119691+8.982761)
#for sample C (1/n*(c,a)+(c,b)+(c,d))
C=(1/3)*(7.211103+7.119691+3.000000)
#for sample D (1/n*(d,a)+(d,b)+(d,c))
D=(1/3)*(9.219544+8.982761+3.000000)
print(A,B,C,D)
```

#### Sample D is more dissimilar (More distance) from others.So now we going to create two clusters.Ci={a,b,c,d} and Cj=.Now we take sample D and put it to Cj

```python
Ci={A,B,C}
Cj={D}
```

#### Now compute dissimilar of x in Ci and x in Cj

```python
#for sample A to Ci and Cj
A=((1/2)*(0.538516+7.211103))-((1/1)*(9.219544))
#for sample B to Ci and Cj
B=((1/2)*(0.538516+7.119691))-((1/1)*(8.9827))
#for sample C to Ci and Cj
C=((1/2)*(7.211103+7.119691))-((1/1)*(3.000000))
print(A,B,C)
```

#### Sample C is more dissimilar (More distance) from others.So the clusters become Ci={a,b} and Cj={{d}c}.Now compute dissimilarity

```python
dist = np.sqrt(((df.values[:, None, :] - df.values[None, :, :]) ** 2).sum(axis=2))

Z = linkage(dist, method='ward')
plt.figure(figsize=(6, 4))
dendrogram(Z, labels=df.index.tolist())
plt.tight_layout()
plt.show()
```
