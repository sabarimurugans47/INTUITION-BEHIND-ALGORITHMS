# **Importing necessary libraries**

```python
import pandas as pd,matplotlib.pyplot as plt,numpy as np
```

# **Creating a DataFrame**

```python
X=np.array([2,5,8,9,10])
Y=np.array([4,11,15,17,19])
```

```python
Data=pd.DataFrame({'X':X,'Y':Y})
```

```python
Data
```

# **Building Simple Linear Regression Function**

```python
def simple_linear_regression(x):
    X_mean=np.mean(X)
    Y_mean=np.mean(Y)
    X_Y=X*Y
    X2=X**2
    coeff = (np.mean(X_Y) - X_mean * Y_mean) / (np.mean(X2) - X_mean**2)
    bias=Y_mean-(coeff*X_mean)
    y=bias+(coeff*x)
    return coeff,bias,y
simple_linear_regression(10)
```

# **Predicting using manually computed function**

```python
print(simple_linear_regression(2))
```

The predicted value for input 2 is 4.54205607476635

# **Fitting inbuilt Linear Regression class in sklearn**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(np.reshape(X,(-1,1)),np.reshape(Y,(-1,1)))
```

```python
lr.coef_,lr.intercept_
```

# **Plotting**

```python
plt.scatter(X,Y)
plt.plot(X,Y)
plt.title('Actual Data')
plt.show()
```

```python
plt.scatter(X,Y)
plt.plot(X,lr.predict(np.reshape(X,(-1,1))))
plt.title('Predicted Data')
plt.show()
```

# **Spotting Intercept**

```python
X = np.array([2, 5, 8, 9, 10])
Y = np.array([4, 11, 15, 17, 19])

X_mean = np.mean(X)
Y_mean = np.mean(Y)
coeff = (np.mean(X * Y) - X_mean * Y_mean) / (np.mean(X**2) - X_mean**2)
bias = Y_mean - coeff * X_mean

plt.scatter(X, Y, label='Data points')
plt.plot(X, bias + coeff * X, color='green', label='Regression line')
plt.scatter(0, bias, color='red', s=100, label='Bias point (Y-intercept)')
plt.annotate('Bias (Y-intercept)',
             xy=(0, bias),
             xytext=(2, bias + 5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=12)
plt.title('Regression Line with Bias Point')
plt.show()
```
