```python
import pandas as pd,numpy as np
```

```python
data = {
    'age': [23, 31, 35, 35, 42, 43, 45, 46, 46, 51],
    'likes_goats': [0, 1, 0, 0, 0, 1, 1, 1, 0, 1],
    'likes_height': [0, 1, 1, 0, 0, 1, 0, 1, 1, 1],
    'climbed_meters': [200, 700, 600, 300, 200, 700, 300, 700, 600, 700]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['Initial Predictions'] = df['climbed_meters'].mean()
df
```

```python
df['Residuals']=df['climbed_meters']-df['Initial Predictions']
```

## Training Tree 1

```python
group_A=df[df['likes_height']==1] #Leaf 1
group_B=df[df['likes_height']==0] #Leaf 2
```

```python
group_A
```

```python
group_B
```

```python
Leaf_A=group_A['Residuals'].mean()
Leaf_B=group_B['Residuals'].mean()
Leaf_A,Leaf_B
```

```python
df.drop('Residuals',axis=1,inplace=True)
```

## Updation Formula : Avg + (0.1 * Tree 1 predictions)

```python
s1=500+(0.1*166.66666666666666)
s2=500+(0.1*166.66666666666666)
s5=500+(0.1*166.66666666666666)
s7=500+(0.1*166.66666666666666)
s8=500+(0.1*166.66666666666666)
s9=500+(0.1*166.66666666666666)
```

```python
s0=500+(0.1*-250.0)
s3=500+(0.1*-250.0)
s4=500+(0.1*-250.0)
s6=500+(0.1*-250.0)
df['tree1']=pd.Series([s0,s1,s2,s3,s4,s5,s6,s7,s8,s9])
df
```

```python
df['Residual']=df['climbed_meters']-df['tree1']
df
```

## Training Tree 2

```python
tree2=df[['age','likes_goats','likes_height','Residual']]
tree2
```

```python
group_C = df[df['likes_goats'] == 1]
group_D = df[df['likes_goats'] == 0]

leaf_C = group_C['Residual'].mean()
leaf_D = group_D['Residual'].mean()
print("Leaf C:", leaf_C, "Leaf D:", leaf_D)
```

```python
lr = 0.1
df['tree2_output'] = df['likes_goats'].apply(lambda x: lr * leaf_C if x == 1 else lr * leaf_D)

df['tree2'] = df['tree1'] + df['tree2_output']
df[['climbed_meters', 'tree1', 'tree2', 'Residual']]
df
```

```python
df['tree2 residuals']=df['Residual']-df['tree2']
df
#
```
